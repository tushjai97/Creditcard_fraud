```python
from sklearn.preprocessing import RobustScaler
robust = RobustScaler()
dataset['scaled_time'] = robust.fit_transform(dataset['Time'].values.reshape(-1,1))
dataset['scaled_money'] = robust.fit_transform(dataset['Amount'].values.reshape(-1,1))
```

```python
col_drop = ['Time','Amount']
col_add = ['scaled_time','scaled_money']
```

```python
def drop_columns(df,columns):
  df = df.copy()
  col = []
  col = columns
  df.drop(col, axis=1, inplace=True)
  return df
```

```python
def add_col(df,columns):
  df = df.copy()
  col = []
  col = columns
  for item in col:
    df[item]
  return df
```

```python
dataset_without_time_amount = drop_columns(dataset,col_drop)
dataset_standardized = add_col(dataset_without_time_amount,col_add)
```

Undersampled dataset with Class 0 = 492 and Class 1 = 492

```python
#Randomizing the dataset
df = dataset_standardized
df = df.sample(frac=1)

fraud_data = df.loc[df['Class']==1]
non_fraud_data = df.loc[df['Class']==0][:492]

undersampled_data = pd.concat([fraud_data,non_fraud_data])
undersampled_data = undersampled_data.sample(frac=1, random_state=42)
print(undersampled_data['V14'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

classifiers = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
    "Support Vector Classifier": SVC(),
    "DecisionTreeClassifier": DecisionTreeClassifier()
}
```

```python
from sklearn.model_selection import train_test_split, cross_validate, cross_val_score
for key,classifier in classifiers.items():
  classifier.fit(X_train_undersampled,y_train_undersampled)
  training_score = cross_val_score(classifier, X_train_undersampled,y_train_undersampled, cv =5)
  print("Classifiers: ", classifiers[key], "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy score")
```

```python
from sklearn.model_selection import GridSearchCV
log_reg_param = {'penalty': ['l1','l2'], 
                 'C': [0.001,0.01,0.1,1,10,100]}
log_reg_grid = GridSearchCV(LogisticRegression(solver='liblinear'), log_reg_param)
log_reg_grid.fit(X_train_undersampled,y_train_undersampled)
log_reg = log_reg_grid.best_estimator_

Knear_param = {
    'n_neighbors': list(range(2,5,1)), 
     'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],

}
knear_grid = GridSearchCV(KNeighborsClassifier(), Knear_param)
knear_grid.fit(X_train_undersampled,y_train_undersampled)
knear = knear_grid.best_estimator_

SVC_param = {
     'C': [0.5, 0.7, 0.9, 1], 
     'kernel': ['rbf', 'poly', 'sigmoid', 'linear'],
    #  'probability' : ['True']
     }

SVC_grid = GridSearchCV(SVC(), param_grid = SVC_param)
SVC_grid.fit(X_train_undersampled,y_train_undersampled)
svc = SVC_grid.best_estimator_

print("Best Logistic Regression", log_reg)
print("Best Knear", knear)
print("Best SVC", svc)



```

```python
log_reg_score = cross_val_score(log_reg, X_train_undersampled,y_train_undersampled, cv=5)
print("Logistic Regression accuracy: ", round(log_reg_score.mean(),3)*100, "%")
knear_score = cross_val_score(knear,X_train_undersampled,y_train_undersampled, cv=5)
print("Knearest neighbor accuracy: ", round(knear_score.mean(),2)*100,"%")
svc_score = cross_val_score(svc,X_train_undersampled,y_train_undersampled, cv=5)
print("SVC accuracy: ", round(svc_score.mean(),3)*100,"%")
```
